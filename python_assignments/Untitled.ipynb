# decorators

```python
# decorators in python 
def decorator(func):
    def wrapper():
        print("before")
        func()
        print("after")
    return wrapper 
@decorator
def greet():
    print("hello, world!")
greet()    
```

```python
def decorator_name(func): 
    def wrapper(*args,**kwargs):
        print("before")
        result = func(*args, **kwargs)
        print("after")
        return result 
    return wrapper 
@decorator_name
def add(a,b):
    return a+b
print(add(5,3))    
    
```

# generators

```python
def my_generator():
    yield 1
    yield 2
    yield 3 
for value in my_generator():
    print(value)
```
