```python
import numpy as np

def logistic_regression_gd(X, Y, k, tau, lam):
    
    n, m = X.shape
    
    # Add intercept column
    X = np.hstack((np.ones((n, 1)), X))
    
    # Initialize beta randomly (m+1 x 1)
    beta = np.random.randn(m + 1, 1)
    
    prev_cost = float('inf')
    
    for i in range(k):
        
        # Linear combination
        z = np.dot(X, beta)
        
        # Sigmoid
        y_hat = 1 / (1 + np.exp(-z))
        
        # Cost function
        cost = -(1/n) * np.sum(
            Y * np.log(y_hat + 1e-10) + 
            (1 - Y) * np.log(1 - y_hat + 1e-10)
        )
        
        # Check stopping condition
        if abs(prev_cost - cost) < tau:
            break
        
        # Gradient
        gradient = (1/n) * np.dot(X.T, (y_hat - Y)) # ∂J/∂β=(1/n)XT(y^​−Y)
        
        # Update beta
        beta = beta - lam * gradient
        
        prev_cost = cost
    
    return beta, cost
```
