```python
import numpy as np 
```

```python
def generate_logistic_data(theta,n,m):
    beta = np.random.randn(m + 1)# generate random numbers from normal distribution(mean = 0, std = 1)
    X_random = np.random.randn(n, m)
    ones = np.ones((n, 1))
    X = np.hstack((ones, X_random))
    z = np.dot(X, beta)
    p = 1 / (1 + np.exp(-z))
    Y = (p > 0.5).astype(int)
    flip = np.random.binomial(1, theta, n)
    Y = np.abs(Y - flip)
    Y = Y.reshape(-1, 1)
    return X, Y, beta
```

```python
theta = 0.2   # 20% noise
n = 100       # 100 samples
m = 3         # 3 independent variables

X, Y, beta = generate_logistic_data(theta, n, m)

print("X shape:", X.shape) # X в†’ input matrix
print("Y shape:", Y.shape) # Y в†’ binary labels (output)
print("Beta:", beta) # true coefficients
```
